```python
# Group the data by Region and calculate the total Sales for each region
total_sales_by_regin = sales.groupby('Region')['Sales'].sum().reset_index()
print("Total Sales by Region:\n", total_sales_by_regin)
```

```python
# Group by Region and Product, then calculate the total Sales and average Units
total_sales_avg_units_by_region_product = sales.groupby(['Region','Product']).agg({'Sales':'sum',"Units":'mean'}).reset_index()
print("\nTotal Sales and Average Units by Region and Product:\n", total_sales_avg_units_by_region_product)
```

```python
# Group by Product and calculate multiple aggregations for Sales and Units
aggregations_by_product = sales.groupby('Product').agg({
    'Sales':['sum','mean','max'],
    'Units':['sum','mean','max']
}).reset_index()
aggregations_by_product
```
