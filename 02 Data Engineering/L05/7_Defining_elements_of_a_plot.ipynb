### Defining axes, ticks, and grids

```python
# ax = fig.add_axes([.1,.1,5,5])
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure() # figure 是一个空白的画布，图表空间本身
ax = fig.add_axes([1,1,1,1]) # axe 是坐标系， axi 是坐标轴，add_axes 是在画布上添加坐标系, 参数表示的意义：[left, bottom, width, height]，取值范围是0-1，表示占画布的比例
ax.plot(x,y)
```

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]
fig = plt.figure()
ax =fig.add_axes([0.5,0.5,2,2])
ax.plot(x,y)
```

```python
# fig = plt.figure()
# ax = fig.add_axes([.1,.1,1,1])
#
# ax.set_xlim([1,9])
# ax.set_ylim([0,5])
#
# ax.set_xticks([0,1,2,4,5,6,8,9,10])
# ax.set_yticks([0,1,2,3,4,5])
#
# ax.plot(x,y)

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10]) # 缺少3,7
ax.set_yticks([0,1,2,3,4,5]) # 全部显示

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots

```python
# #fig, (ax1, ax2) = plt.subplots(2,1)
# fig = plt.figure()
# fig, (ax1,ax2) = plt.subplots(1,2)
# ax1.plot(x)
# ax2.plot(x,y)

fig = plt.figure()
fig, axs = plt.subplots(1,2) # 1 row, 2 columns
axs[0].plot(x)
axs[1].plot(x,y)
```

```python
fig = plt.figure()
fig, axs = plt.subplots(2,2)
#fig, (ax1, ax2) = plt.subplots(1,2)
axs[0,0].plot(x)
axs[0,1].plot(x,y)
axs[1,0].plot(x)
axs[1,1].plot(x,y)
```
