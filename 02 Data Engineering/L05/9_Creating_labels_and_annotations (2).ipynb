## Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)

plt.xlabel('This is the x-axis')
plt.ylabel('This is the y-axis')

plt.show()
```

```python
# z = [1,2,3,4,.5]
# veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']
#
# plt.pie(z, labels=veh_type)
# plt.show()

z = [1,2,3,4,.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z,labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
cars = pd.read_csv('./mtcars.csv')
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = cars['mpg']

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')
```

```python
plt.figure(figsize=(8,4))
plt.plot(mpg)
plt.xlabel('car names')
plt.ylabel('miles/gal')
plt.title('Miles per Gallon of Cars in mtcars Dataset')
plt.xticks(ticks=range(32), labels=cars.car_names, rotation=60, fontsize='medium')
plt.show()
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
 # try to work with 'loc' from matplotlib documentation and try other way
```

### Annotating your plot

```python
mpg.max()
```

```python
mpg.argmax()
```

```python
cars.car_names[mpg.argmax()]
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')

ax.set_ylim([0,45])

ax.annotate(cars.car_names[mpg.argmax()], xy= (19,33.9), xytext=(21,35),
           arrowprops=dict(facecolor='black', shrink=0.05))
# annotate the lowest mpg (argmin) car in same figure
```

```python
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.legend(loc='best')
ax.set_ylim([0,45])

ax.annotate(cars.car_names[mpg.argmin()], xy=(14,10.4), xytext=(5,5),
           arrowprops=dict(facecolor='red', shrink=0.05))
```
