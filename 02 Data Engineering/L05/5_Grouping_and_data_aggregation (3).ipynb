# Class exercise:
Load the dataset sales_data.csv into a DataFrame.

Group the data by Region and calculate the total Sales for each region.

Group by Region and Product, then calculate the total Sales and average Units.

Group by Product and calculate multiple aggregations for Sales and Units

```python
sales = pd.read_csv('./sales_data.csv')
sales.head()
```

```python
sales_region = sales.groupby('Region').sum(numeric_only=True)
sales_region
```

```python
sales_product = sales.groupby('Product')
sales_product_sum = sales_product.sum(numeric_only=True)
sales_product_sum
```

```python
sales_product_avg = sales_product.mean(numeric_only=True)
sales_product_avg
```

```python
sales_product_std = sales_product.std(numeric_only=True)
sales_product_std
```
