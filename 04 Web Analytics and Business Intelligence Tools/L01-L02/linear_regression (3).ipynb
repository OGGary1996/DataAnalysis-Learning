```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

# homeprice = m*area + C

```python
df = pd.read_csv('homeprices.csv')
df
```

what is price of home if we have 5000 sqm

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(8,6))
ax.set_title("Home Prices vs Area")
ax.set_xlabel("Area (sqr ft)")
ax.set_ylabel("Price (in $)")
ax.scatter(df["area"], df["price"], color='red', marker='+')
plt.show()
```

```python
# #seperating the x/independent variable
# x input 必须保持二维的格式
# 可能存在多个输入变量 x1, x2, x3...
# y output 可能只有一个变量

area = df[["area"]] # input 不能直接用 df.area, 因为它是一维的 Series
area
```

```python
price = df.price  #y /dependent variable
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(area,price)  #.fit =>training (input,output)

```

**(1) Predict price of a home with area = 3300 sqr ft**

```python
reg.predict([[3300]])
```

```python
reg.coef_   #slope of line
```

```python
reg.intercept_   #y-axis intercept
```
